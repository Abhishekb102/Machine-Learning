# Performing logistic regression

# Process the dataset

```python
df['class'].value_counts(normalize=1)*100
```

```python
from sklearn.model_selection import train_test_split
```

```python
# create training and testing vars
x_train, x_test, y_train, y_test=train_test_split(df[['mcg', 'gvh', 'alm', 'mit', 'erl', 'pox', 'vac', 'nuc']],df['class'], test_size=0.03,random_state=2182,stratify=df['class'])
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

# Using logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predicted_class=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,predicted_class)
```

```python
accuracy
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predicted_class))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test, predicted_class)
```

```python
cm
```

# Using neural network

```python
from sklearn.neural_network import MLPClassifier
```

```python
nnmodel=MLPClassifier(hidden_layer_sizes=(30,40))
```

```python
nnmodel.fit(x_train,y_train)
```

```python
nnpredicted_class=nnmodel.predict(x_test)
```

```python
accuracy_score(y_test,nnpredicted_class)
```

```python
print(classification_report(y_test,nnpredicted_class))
```

```python
nn_cm=confusion_matrix(y_test, nnpredicted_class)
```
